```python
import sympy as sp
import numpy as np
import math
from matplotlib import pyplot as plt
x = sp.symbols('x')

y = sp.integrate(x**2 + 10*sp.sin(x), (x, -10, 10))
#print(y)
print("Numerical integration result:", float(y))
x = np.linspace(-10,10,100)
y = x**2 + 10*np.sin(x)
plt.title('Area under the graph of $x^2 + 10 \sin(x)$',fontsize = 'xx-large',fontstyle='normal',fontweight='medium')
plt.ylabel('',fontsize = 'xx-large',fontstyle='oblique',fontweight = 'heavy')
plt.xlabel('x',fontsize = 'xx-large',fontstyle='normal',fontweight = 'medium')
plt.xlim(-10,10)
plt.ylim([min(y), max(y)])
plt.fill_between(x, y,color= "b",alpha= 0.4)
plt.plot(x,y,'r')
plt.show()

```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.integrate import quad

def func(x):
    return x**2 + 10*np.sin(x)

result = quad(func, -10, 10)

x = np.linspace(-10, 10, 100)
y=func(x)
plt.title('Area under the graph of $x^2 + 10 \sin(x)$',fontsize = 'xx-large',fontstyle='normal',fontweight='medium')
plt.ylabel('',fontsize = 'xx-large',fontstyle='oblique',fontweight = 'heavy')
plt.xlabel('x',fontsize = 'xx-large',fontstyle='normal',fontweight = 'medium')
plt.xlim(-10,10)
plt.ylim([min(y), max(y)])
plt.fill_between(x, y,color= "b",alpha= 0.4)
plt.plot(x, y, 'r')

print("Numerical integration result:",result)
```

```python
#finding extrema, scipy
```

```python
import numpy as np
import math
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

def func(x):
    return x**2 + 10*np.sin(x)

result = minimize_scalar(func)
print("Minimum value:", result.fun)
print("Location of minimum:", result.x)

x = np.linspace(-10, 10, 100)
y=func(x)
plt.title('Area under the graph of $x^2 + 10 \sin(x)$',fontsize = 'xx-large',fontstyle='normal',fontweight='medium')
plt.ylabel('',fontsize = 'xx-large',fontstyle='oblique',fontweight = 'heavy')
plt.xlabel('x',fontsize = 'xx-large',fontstyle='normal',fontweight = 'medium')
plt.xlim(-10,10)
plt.ylim([min(y), max(y)])
plt.fill_between(x, y,color= "b",alpha= 0.4)
plt.plot(x, y, 'r')

#print("Numerical integration result:",float(y))
```

```python
#Integration Using Sympy
```

```python
import sympy as sp

x = sp.symbols('x')

expr = x**2 + 10*sp.sin(x) 

integral_symb = sp.integrate(expr, x)

definite_integral = sp.integrate(expr, (x, -10, 10))

print("Symbolic integral:", integral_symb)
print("Definite integral from -10 to 10:", definite_integral.evalf())
```
