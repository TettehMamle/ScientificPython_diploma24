```python
from scipy import optimize
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
np.random.seed(0)


def f(x, a, phi):
    return np.cos(a * x + phi)

x = np.linspace(-2, 4, 50)
y = f(x, 1.3, 0.9) + .1*np.random.normal(size=50)
np.savetxt("saved_test.txt",np.column_stack([x,y]),fmt='%f') #saving to a file



my_data=np.genfromtxt('saved_test.txt',dtype=float) #reading from a file
params, params_cov = optimize.curve_fit(f,my_data[:,0], my_data[:,1])

print(params) # prints the coordinates after running
plt.title('Fitting a Function',fontsize = 'xx-large',fontstyle='normal',fontweight='medium')
plt.plot(my_data[:,0],my_data[:,1] , 'bx',label='data')
plt.plot(x,f(x,1.3,0.9),'g-',label="Original function \na=1.3, phi=0.9")
plt.ylabel('cos(a * x + phi)',fontsize = 'xx-large',fontstyle='oblique',fontweight = 'medium')
plt.xlabel('x-axis',fontsize = 'xx-large',fontstyle='normal',fontweight = 'medium')
plt.plot(my_data[:,0], f(my_data[:,0], *params), 'r-', label="Fitting \na=1.27, phi=0.93")
plt.legend(loc="upper left")
```
