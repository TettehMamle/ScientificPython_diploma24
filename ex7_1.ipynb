```python
from scipy import optimize
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(0)


def f(x, a, phi):
    return np.cos(a * x + phi)

x = np.linspace(0, 4, 50)
y = f(x, 1.3, 0.9) + .1*np.random.normal(size=50)
np.savetxt("saved_test.txt",np.column_stack([x,y]),fmt='%f')



my_data=np.genfromtxt('saved_test.txt',dtype=float)
params, params_cov = optimize.curve_fit(f,my_data[:,0], my_data[:,1])

print(params)
plt.plot(my_data[:,0],my_data[:,1] , 'bx')
plt.plot(x,f(x,1.3,0.9),'g-',label="Origin")
plt.ylabel('sin',fontsize = 'xx-large',fontstyle='oblique',fontweight = 'heavy')
plt.xlabel('x-axis',fontsize = 'xx-large',fontstyle='normal',fontweight = 'medium')
plt.plot(my_data[:,0], f(my_data[:,0], *params), 'r-', label="Fitting")
plt.legend(loc="upper left")
```
