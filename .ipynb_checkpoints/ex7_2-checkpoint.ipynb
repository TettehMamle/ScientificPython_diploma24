```python
import sympy as sp
import numpy as np
import math
from matplotlib import pyplot as plt
x = sp.symbols('x')

y = sp.integrate(x**2 + 10*sp.sin(x), (x, -10, 10))
print(y)

x = np.linspace(-10,10,100)
y = x**2 + 10*np.sin(x)
plt.ylabel('x**2 + 10*sin(x)',fontsize = 'xx-large',fontstyle='oblique',fontweight = 'heavy')
plt.xlabel('x',fontsize = 'xx-large',fontstyle='normal',fontweight = 'medium')
plt.xlim(-10,10)
plt.fill_between(x, y,
color= "g",
alpha= 0.4)
plt.plot(x,y)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import minimize_scalar

def func(x):
    return x**2 + 10*np.sin(x)

result = minimize_scalar(func)
print("Minimum value:", result.fun)
print("Location of minimum:", result.x)

x = np.linspace(-10, 10, 100)
y=func(x)
plt.ylabel('x**2 + 10*sin(x)',fontsize = 'xx-large',fontstyle='oblique',fontweight = 'heavy')
plt.xlabel('x',fontsize = 'xx-large',fontstyle='normal',fontweight = 'medium')
plt.xlim(-10,10)
plt.fill_between(x, y,
color= "g",
alpha= 0.4)
plt.plot(x, y, 'b-')
```
